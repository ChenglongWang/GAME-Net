Here, we apply the target scaling with the target_scale function, providing the optional parameter mode="std" in order to apply standardization. Normalization can be applied instaed, just providing mode="norm".

## GNN Training

### Optimizer

Used optimizer for the training is Adam, algorithm for first-order gradient-based optimization of
stochastic objective functions, based on adaptive estimates of lower-order mo-
ments.

```python
lr = 0.001  # Initial learning rate
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

### Learning Rate (LR) Scheduler

Helps steering the learning rate during the training, providing faster convergence and higher accuracy. The used scheduler is the "Reduce On Loss Plateau Decay".

```python
patience = 5   # number of epochs with no improvement after which lr will be reduced (original is 5)
factor = 0.7   # multiplier to decrease the lr
min_lr = 1e-5  # minimum learning rate
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer,
                                                       mode='min',
                                                       factor=factor,
                                                       patience=patience,
                                                       min_lr=min_lr)
```

### Run Training

```python
best_val_error = None
loss_list = []  # Store loss function trend during training
val_list = []   # Store validation score during training
epochs = 300    # Number of epochs 

t0 = time.time()
for epoch in range(1, epochs+1):
    lr = scheduler.optimizer.param_groups[0]['lr']
    loss = train_loop(model, epoch, device, train_loader, optimizer)  # Run epoch and update params based on loss
    val_error = test_loop(model, val_loader, device, std_tv)          # Run epoch on validation set
    scheduler.step(val_error)                                         # Adjust LR according to validation error

    if best_val_error is None or val_error <= best_val_error:         # If val_error improves, update test_error
        test_error = test_loop(model, test_loader, device, std_tv)
        best_val_error = val_error                                    # Update best val_error

    print('Epoch {:03d}: LR={:7f}  Loss={:.6f}  Validation MAE: {:.6f} eV, '
          'Test MAE: {:.6f} eV'.format(epoch, lr, loss, val_error, test_error))

    loss_list.append(loss)
    val_list.append(val_error)
print("----------------------------------------------------------------------------------------")
print("device: {}    Training time: {:.2f} s".format(device, time.time() - t0))
```

### Test set: Performance Analysis

```python
def test_performance(model, test_loader):
    """
    Analyzes performance of the trained model based on the test set.
    Outlier criterion based on error distribution: if the error is higher than 3*std, 
    the corresponding sample is defined as outlier.    
    Args:
        model(Net): GNN model to be analyzed
        test_loader(): test loader 
    Returns:
        Report about model performance (printed)
        

    """
    model.eval()
    w_pred = []
    w_true = []
    label_list  = []
    for batch in test_loader:
        batch = batch.to("cpu")
        w_pred += model(batch)
        w_true += batch.y
    N = len(test_loader.dataset)
    for data in test_loader.dataset:
        label_list.append(get_graph_formula(data, encoder.categories_[0]))
    w_pred = [i.item() for i in w_pred]
    w_true = [i.item() for i in w_true]
    y_pred = [w_pred[i]*std_tv + mean_tv for i in range(N)]
    y_true = [w_true[i]*std_tv + mean_tv for i in range(N)]
    # Error arrays
    E = [(y_true[i] - y_pred[i]) for i in range(N)]
    AE = [abs(E[i]) for i in range(N)]
    SE = [E[i]**2 for i in range(N)]
    APE = [abs(E[i]/y_true[i]) for i in range(N)] # Absolute Percentage Error
    error_dict = dict(zip(label_list, E))
    abs_error_dict = dict(zip(label_list, AE))
    square_error_dict = dict(zip(label_list, SE))
    # Error distirbutions
    sns.displot(E, bins=50, kde=True)
    plt.tight_layout()
    plt.show()
    plt.close()
    sns.displot(AE, bins=50, kde=True)
    plt.tight_layout()
    plt.show()
    plt.close()
    # Performance metrics
    ME = np.mean(E)        # eV
    MAE = np.mean(AE)      # eV
    MSE = np.mean(SE)      # eV^2
    RMSE = np.sqrt(MSE)    # eV
    MAPE = np.mean(APE) * 100 # %
    s_E = np.std(E)        # eV
    s_AE = np.std(AE)      # eV
    s_APE = np.std(APE)    # %
    print("ME = {:.2f} eV    MAE = {:.2f} eV    RMSE = {:.2f} eV    MAPE = {:.2f}%".format(ME, MAE, RMSE, MAPE))
    print("sE = {:.2f} eV    sAE = {:.2f} eV    sAPE = {:.2f} %".format(s_E, s_AE, s_APE))
    print("------------------------------------------------------------------")
    # Outliers detection
    outliers_list = []
    outliers_error_list = []
    index_list = []
    counter = 0
    for i in range(N):
        if abs(E[i]) >= 3 * s_E:  # Outlier criterion based on E, define the border!
            counter += 1
            outliers_list.append(label_list[i])
            outliers_error_list.append(E[i])
            index_list.append(i)
            if counter < 10:
                print("0{}) Outlier Detected: {}    Error: {:.2f} eV    (index={})".format(counter,
                                                                                 label_list[i],
                                                                                 E[i], 
                                                                                 i))
            else:
                print("{}) Outlier Detected: {}    Error: {:.2f} eV    (index={})".format(counter,
                                                                                label_list[i],
                                                                                E[i], 
                                                                                i))
            plotter(test_loader.dataset[i])
            plt.show()
            plt.close()
    outliers_dict = dict(zip(outliers_list, outliers_error_list))
    return abs_error_dict, outliers_dict
    
```
